```python
import os
from PIL import Image

directorio = 'DATABASE/darknet/labeled_images'

def cambiar_jpeg_a_jpg(directorio):
    archivos = os.listdir(directorio)
    # Iterar sobre cada archivo en el directorio
    for archivo in archivos:
        if archivo.endswith('.jpeg'):  # Verificar si es un archivo .jpeg
            # Construir la ruta completa del archivo
            ruta_completa = os.path.join(directorio, archivo)
            
            # Abrir la imagen usando PIL
            imagen = Image.open(ruta_completa)
            
            # Crear un nuevo nombre de archivo reemplazando la extensión .jpeg por .jpg
            nuevo_nombre = os.path.splitext(ruta_completa)[0] + '.jpg'
            
            # Guardar la imagen en formato JPEG con el nuevo nombre
            imagen.save(nuevo_nombre, 'JPEG', quality=95)  # Puedes ajustar la calidad según necesites
            
            # Cerrar la imagen abierta
            imagen.close()
            
            # Eliminar el archivo original (.jpeg)
            os.remove(ruta_completa)


'''
Previamente --> 0 <x_min> <y_min> <x_max> <y_max>
Ahora --> 0 <x_center> <y_center> <width> <height>
'''
def cambiar_estructura_txt(directorio):
    archivos = os.listdir(directorio)

    for archivo in archivos:
        if archivo.endswith('.txt') and archivo != 'classes.txt':  
            ruta_etiquetas = os.path.join(directorio, archivo)
            
            with open(ruta_etiquetas, 'r') as file:
                lineas = file.readlines()
            
            with open(ruta_etiquetas, 'w') as new_file:
                for linea in lineas:
                    # Separar las coordenadas en una lista de valores
                    valores = linea.split()
                                        
                    # Ignorar el primer valor (clase del objeto) y convertir los restantes a números flotantes
                    try:
                        x_min, y_min, x_max, y_max = map(float, valores[1:])
                    except ValueError:
                        print(f"Error: No se pudieron convertir los valores en la línea '{linea.strip()}' de '{archivo}'.")
                        # Escribir la línea original de vuelta al archivo
                        new_file.write(linea)
                        continue  # Saltar esta línea si no se pueden convertir los valores
                    
                    # Calcular las coordenadas en x_center, y_center, width, height
                    x_center = (x_min + x_max) / 2
                    y_center = (y_min + y_max) / 2
                    width = x_max - x_min
                    height = y_max - y_min
                    
                    # Escribir las nuevas coordenadas en el nuevo archivo
                    new_line = f"{x_center} {y_center} {width} {height}\n"
                    new_file.write(new_line)
            
            print(f"Convertido: {archivo}")

    print("¡Proceso completado!")
```

```python
cambiar_jpeg_a_jpg(directorio)
```

```python
cambiar_estructura_txt(directorio)
```

```python
def agregar_cero_en_inicio(directorio):
    # Obtener la lista de archivos en el directorio
    archivos = os.listdir(directorio)

    # Iterar sobre cada archivo en el directorio
    for archivo in archivos:
        if archivo.endswith('.txt') and archivo != 'classes.txt':  # Verificar si es un archivo de etiquetas
            # Construir la ruta completa del archivo de etiquetas
            ruta_etiquetas = os.path.join(directorio, archivo)
            
            # Leer las líneas del archivo de etiquetas
            with open(ruta_etiquetas, 'r') as file:
                lineas = file.readlines()
            
            # Abrir un nuevo archivo para escribir las etiquetas modificadas
            with open(ruta_etiquetas, 'w') as new_file:
                for linea in lineas:
                    # Agregar un '0' al inicio de cada línea
                    nueva_linea = f"0 {linea}"
                    new_file.write(nueva_linea)
            
            print(f"Agregado '0' a cada línea en: {archivo}")

    print("¡Proceso completado!")
```

```python
agregar_cero_en_inicio(directorio)
```
