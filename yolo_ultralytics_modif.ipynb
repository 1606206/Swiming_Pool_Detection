## **Hacemos la prediccion  para todas las imagenes de test**
**Primero  lo haremos con un entrenamiento de 10 epochs**

```python
from pathlib import Path
# Rutas de directorios
input_dir = Path('DATABASE/ultralytics_modif/images/test')
output_dir_10_epoch = Path('DATABASE/ultralytics/images/10epoch_accuracy')

# Verificar si los directorios de salida existen, si no, crearlos
output_dir_10_epoch.mkdir(parents=True, exist_ok=True)

model_10_epoch =YOLO("ultralytics_trainings/10epoch_accuracy_modif.pt")
image_files = list(input_dir.glob('*.*'))

accuracys_10epoch = []
for image_file in image_files:
    filename = image_file.stem

    output_path_10_epoch = output_dir_10_epoch / f"{filename}.jpg"
    results_10_epoch = model_10_epoch([str(image_file)])

    
    for result in results_10_epoch:
        boxes = result.boxes
        if len(boxes) > 0:
            for box in boxes:
                acc = tf.concat(box.conf, axis=0).numpy()
                accuracys_10epoch = np.append(accuracys_10epoch, acc)
        else:
            acc = tf.concat(boxes.conf, axis=0).numpy()
            accuracys_10epoch = np.append(accuracys_10epoch, acc)
        result.save(filename=str(output_path_10_epoch))
```

**Finalmente con 100 epoch**

```python
from pathlib import Path
# Rutas de directorios
input_dir = Path('DATABASE/ultralytics/images/test')
output_dir_100_epoch = Path('DATABASE/ultralytics/images/100epoch_accuracy')

# Verificar si los directorios de salida existen, si no, crearlos
output_dir_100_epoch.mkdir(parents=True, exist_ok=True)

model_100_epoch = YOLO("ultralytics_trainings/best.pt")
image_files = list(input_dir.glob('*.*'))

dict_100_epoch = {}
for image_file in image_files:
    coords_100epoch = []
    filename = image_file.stem

    output_path_100_epoch = output_dir_100_epoch / f"{filename}.jpg"

    results_100_epoch = model_100_epoch([str(image_file)])

    
    for result in results_100_epoch:
        boxes = result.boxes         
        if len(boxes) > 1:
            for box in boxes:
                coord = box.xyxy.tolist()
                coords_100epoch.append(coord)
        elif len(boxes) == 1:
            coord = boxes.xyxy.tolist()
            coords_100epoch.append(coord)
        result.save(filename=str(output_path_100_epoch))

    dict_100_epoch[f"{filename}.jpg"] = coords_100epoch

print(dict_100_epoch)
```
